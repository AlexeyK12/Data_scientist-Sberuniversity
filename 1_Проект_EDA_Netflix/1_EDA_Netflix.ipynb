### Какие оценки пользователи ставят чаще?

- Постройте гистограмму оценок.
- Почему именно такие оценки ставят чаще всего?

```python
data['user_rating_score'].hist(figsize=(12, 5), bins=20, edgecolor='black')
plt.title('Гистограмма оценок пользователей', fontsize=15)
plt.xlabel('оценка')
plt.ylabel('количество')
plt.show()
```

```python
# добавим категорию оценки, чтобы выявить у каких шоу какие баллы
data['rating_score_category'] = pd.cut(data['user_rating_score'], bins=5)
data['rating_score_category'].value_counts()
```

```python
# категории и группы с самым большим количеством оценок
(data.groupby(['rating_score_category', 'rating_level']).
 agg({'rating_level':'count'}).
 rename(columns={'rating_level':'cnt_rating_score'}).
 sort_values(['rating_score_category','cnt_rating_score'], ascending=False)
).query('cnt_rating_score >= 15')
```

### Постройте график, показывающий распределение количества запущенных шоу в зависимости от года.

- Наблюдается ли рост?
- Есть ли выбросы?
- Что из себя представляют выбросы?
- Чем могут быть они вызваны?

```python
plt.figure(figsize=(12, 7))
ax = sns.barplot(x=data['release_year'].value_counts().sort_index().index,
                 y=data['release_year'].value_counts().sort_index(), palette='Greens')
plt.title('Распределение количества запущенных шоу в зависимости от года', fontsize=17)
plt.xlabel('Год')
plt.xticks(rotation=90)
plt.ylabel('Количество')
plt.grid(True, axis='y', linestyle='--')


for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

- наблюдается локальный рост, большой скачёк в 2016 году и сильный спад в 2017

```python
sns.boxplot(x=data['release_year'], color='green')
plt.title('Выбросы количества выпущенных шоу', fontsize=15)
plt.grid()
plt.show()
```

- выбросы представляют из себя количество выпущенных шоу за год меньше 5 до 1997 года;
- это вызвано тем, что компания основана 29 августа 1997 года (https://ru.wikipedia.org/wiki/Netflix);
- возможно, что шоу произведённые с 1940 по 1997 были просто выкуплены компанией

### Сравните среднюю оценку пользователей в 2016 со средней оценкой в 2017.
- Можно ли сделать вывод, что 2017 год успешнее для Netflix? ("Успешнее" значит, что пользователи в среднем ставили более высокие оценки)
- Как еще можно оценить "Успешность"?



```python
print('Средняя оценка пользователей в 2016:',
      round(data.query('release_year == 2016')['user_rating_score'].mean(), 2))

print('Средняя оценка пользователей в 2017:',
      round(data.query('release_year == 2017')['user_rating_score'].mean(), 2))
```

```python
data.query('release_year == 2016')['user_rating_score'].hist(edgecolor='black', label='2016', figsize=(12, 7))
data.query('release_year == 2017')['user_rating_score'].hist(alpha=0.7, edgecolor='black', label='2017', figsize=(12, 7))
plt.title('Распределение оценок пользовательского рейтинга 2016/2017', fontsize=15)
plt.xlabel('оценка')
plt.ylabel('количество')
plt.legend()
plt.show()
```

Поскольку количество оценок в 2016 году на много больше, для сравнения с 2017 корректнее использовать медиану

```python
print('Медианная оценка пользователей в 2016:',
      round(data.query('release_year == 2016')['user_rating_score'].median(), 5))

print('Медианная оценка пользователей в 2017:',
      round(data.query('release_year == 2017')['user_rating_score'].median(), 5))
```

Медианное значение показывает, что 2017 год всего на 1.5 балла успешнее 2016 по оценкам пользователей.

**Наиболее популярные рейтинговые группы выпускаемых шоу:**

- TV-14  -  контент предназначен для зрителей старше 14 лет. Может содержать материал, неприемлемый для детей до 14 лет без присутствия родителей или взрослых;

- TV-MA  -  контент предназначен для взрослых. Этот рейтинг указывает на то, что программа может содержать интенсивные сексуальные сцены, насилие, жестокость, присутствие наркотиков или другие контенты, которые не предназначены для несовершеннолетних;

- PG  -  контент может быть недопустим для детей. Рекомендуется родительское присутствие при просмотре с детьми, так как он может содержать некоторый материал, который может быть не подходящим для младших зрителей;

- G  -  "General Audiences" (общая аудитория). Контент с этим рейтингом подходит для всех возрастов. Он не содержит неприемлемого материала и предназначен для всех зрителей без ограничений.


**Наименее популярные рейтинговые группы выпускаемых шоу:**
- R: рейтинг предназначен для зрителей старше 17 лет. Контент с рейтингом "R" может содержать интенсивные сцены насилия, сексуальные контенты, жестокость, использование наркотиков и другие материалы, которые не рекомендуются для детей и подростков.

- PG-13: "Parental Guidance Suggested" (рекомендуется родительский контроль). Этот рейтинг указывает на то, что некоторый материал в контенте может быть не подходящим для детей до 13 лет. Родительское присутствие и рекомендации родителей рекомендуются для просмотра с детьми.

- NR: "Not Rated" (не оцененный). Этот рейтинг указывает на то, что контент не прошел оценку возрастной категории и его содержание может быть неизвестным или разнообразным. Родителям и зрителям рекомендуется быть более внимательными к контенту с рейтингом "NR", так как он может содержать материалы, которые не подходят для некоторых аудиторий.

- UR: "Unrated" (неоцененный). Этот рейтинг указывает на то, что контент не прошел оценку возрастной категории. Как и с рейтингом "NR", контент с рейтингом "UR" может быть неизвестным или содержать разнообразные материалы, которые могут быть не подходящими для определенных аудиторий.
